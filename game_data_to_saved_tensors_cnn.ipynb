```python
def read_game(game_):
    # all positions with all the moves from a single game
    game_inputs_nn1 = []
    game_inputs_nn2 = []

    game_lables_nn1 = []
    game_lables_nn2 = []

    board = game_.board()

    max_is_white = None
    max_is_playing = None

    if game.headers['White'] == 'maxlich5':
        max_is_white = True
        max_is_playing = True

    elif game.headers['Black'] == 'maxlich5':
        max_is_white = False
        max_is_playing = True

    else:
        max_is_playing = False

    # print('max_is_white:', max_is_white)
    # print('max_is_playing:', max_is_playing)

    for move in game_.mainline_moves():
        if max_is_playing and (board.turn and max_is_white) or (not board.turn and not max_is_white):
            
        
            move_from_square_num = square_parser_full[str(move)[:2]]
            # print(move)
            # print(str(move)[2:])
            # print(str(move))
            # print(square_parser_full[str(move)])
            move_to_square_num = square_parser_full[str(move)[2:]]
            # print(move_to_square_num)

            arrays_input_nn1 = create_whole_input_nn1(board)
            game_inputs_nn1.append(arrays_input_nn1)
            game_lables_nn1.append(move_from_square_num)           


            move_from_array = vectorize_piece(move_from_square_num)
            # print(move_from_array.shape)
            # print(arrays_input_nn1.shape)
            arrays_input_nn2 = np.hstack((arrays_input_nn1, move_from_array))
            game_inputs_nn2.append(arrays_input_nn2)

            game_lables_nn2.append(move_to_square_num)

        
        board.push(move)

    game_inputs_nn1 = np.stack(game_inputs_nn1)
    game_lables_nn1 = np.stack(game_lables_nn1)

    game_inputs_nn2 = np.stack(game_inputs_nn2)
    game_lables_nn2 = np.stack(game_lables_nn2)


    return game_inputs_nn1, game_lables_nn1, game_inputs_nn2, game_lables_nn2
        # print(move)
```

```python
def read_game(g):
    # all positions with all the moves from a single game
    game_inputs = []
    game_lables = []
    board = game.board()
    for move in g.mainline_moves():
        
        # print(board)
        # print(move)
        board.push(move)

        legal_from_squares = from_legal_moves_to_square_numbers(board)

        for s in legal_from_squares:

            game_inputs.append(create_whole_input_nn2(board, s))# .reshape(1,-1)
            game_lables.append(create_y_values_nn2(board, s))# .reshape(1,-1)



    game_inputs = np.stack(game_inputs)
    game_lables = np.stack(game_lables)
    return game_inputs, game_lables
        # print(move)
```

```python
# game lood saving games to games list
pgn_file = open('pgn_files/lichess_maxlich5_2024-02-03.pgn')
game = chess.pgn.read_game(pgn_file)
print(game.headers)
```

```python
# reading first game out of a file
game_inputs, game_lables = read_game(game)
```

```python
# load every game from a file
game = chess.pgn.read_game(pgn_file)
while game is not None:
    
    game_inputs_, game_lables_ = read_game(game)
    print(game_inputs_.shape)
    print(game_inputs.shape)

    game_inputs = np.vstack((game_inputs, game_inputs_))
    game_lables = np.vstack((game_lables, game_lables_))

    print(game_inputs.shape)
    print(game.headers)

    game = chess.pgn.read_game(pgn_file)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(game_inputs, game_lables, test_size=0.2, random_state=33)
```

```python
data_training = torch.tensor(X_train)
data_testing = torch.tensor(X_test)

lables_training = torch.tensor(y_train)
lables_testing = torch.tensor(y_test)
```

```python
torch.save(data_training, 'tensor_data/nn2_rapid_games_data_training_09_02_24.pt')
torch.save(data_testing, 'tensor_data/nn2_rapid_games_data_testing_09_02_24.pt')
```
